```python
import matplotlib.pyplot as plt
import numpy as np
import math
import cv2
```

```python
 # 灰度化
def gray(img_path):
    """
    Calculate function:
    Gray(i,j) = 0.299 * R(i,j) + 0.587 * G(i,j) + 0.114 * B(i,j)
    """
    # 读取图片
    img = plt.imread(img_path)
    # 转换成 RGB 格式
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    # 灰度化
    img_gray = np.dot(img_rgb[...,:3], [0.299, 0.587, 0.114])
        
    return img_gray
```
